```python
data['total_price']=data['total_price'].apply(lambda x:x.replace('万',''))
```

```python
data.head(10)
```

```python
data['total_price'][0]
```

```python
def process(x):
    if '-' in str(x):
        a=str(x).split('-')[0]
        b=str(x).split('-')[1]
        c=(int(a)+int(b))/2
        return int(c)
    else:
        return int(x)
        
data['total_price']=data['total_price'].apply(lambda x:process(x))

```

```python
data.head(10)
```

```python
data['transaction_area']=data['transaction_area'].apply(lambda x:x.replace('平米',''))
```

```python
data.head(10)
```

```python
data['total_price']=data['total_price'].apply(lambda x:int(x))
```

```python
data.groupby('district')['total_price'].sum().sort_values(ascending=False).head(10)
```

```python
data.head(10)
```

```python
data[data['transaction_area']=='--'].index
```

```python
data.index
```

```python
for i in data[data['transaction_area']=='--'].index:
    data.drop([i],axis=0,inplace=True)
```

```python
data[data['transaction_area']=='--'].index
```

```python
data['transaction_area']=data['transaction_area'].apply(lambda x:float(x))
```

```python
data.groupby('district').apply(lambda x:x['transaction_area'].sum()).sort_values(ascending=False).head(10)
```

```python
data['average_price']=data['total_price']/data['transaction_area']*10000
```

```python
data.head(10)
```
